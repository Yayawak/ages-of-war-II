```python
import os
import cv2

def flip_images_in_dir(dir_path):
    # Get a list of all the files and directories in the current directory
    files = os.listdir(dir_path)
    
    # Loop over each file or directory in the current directory
    for file in files:
        # Construct the full path to the file or directory
        full_path = os.path.join(dir_path, file)
        
        # If the file is a directory, recursively call this function on it
        if os.path.isdir(full_path):
            flip_images_in_dir(full_path)
        else:
            # Check if the file is an image with transparency
            if file.endswith(".png") or file.endswith(".tif") or file.endswith(".tiff"):
                img = cv2.imread(full_path, cv2.IMREAD_UNCHANGED)
                img_flipped = cv2.flip(img, 1)
                cv2.imwrite(full_path, img_flipped)

# Call the function on the root directory of your image files
flip_images_in_dir("./")
```
